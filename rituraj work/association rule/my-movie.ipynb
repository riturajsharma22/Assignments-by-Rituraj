# Assignment-09-Association Rules Data Mining (my_movies)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movie=pd.read_csv('../input/rituraj/my_movies.csv')
movie
```

```python
movie.shape
```

```python
movie.info()
```

```python
movie2=movie.iloc[:,5:]
movie2
```

## Apriori Algorithm

### 1. Association rules with 10% Support and 70% confidence

```python
#  10% support
frequent_itemsets=apriori(movie2,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# 70% confidence
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

### 1. Association rules with 5% Support and 90% confidence

```python
# 5% support
frequent_itemsets2=apriori(movie2,min_support=0.05,use_colnames=True)
frequent_itemsets2
```

```python
# 90% confidence
rules2=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.9)
rules2
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules2[rules2.lift>1]
```

```python
plt.scatter(rules2['support'],rules2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
